# Question - 1
plot a line plot between a and b
a = np.arange(40,50)
b = np.arange(50,60)

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
a = np.arange(40,50)
b = np.arange(50,60)
plt.xlabel("a -->", color = "blue")
plt.ylabel("b -->", color = "blue")
plt.title("Line Plot", color = "green")
plt.plot(a,b, "go--", linewidth = 2, markersize = 5)

```

# create a 3 by 3 subplots: multiple plots
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [40,70,90,70]

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [40,70,90,70]
plt.subplot(3,3,1)
plt.plot(x,y1,y2,y3,y4,"r--")
plt.subplot(3,3,2)
plt.plot(x,y1,y2,y3,y4,"g--")
plt.subplot(3,3,3)
plt.plot(x,y1,y2,y3,y4,"b*-")
plt.subplot(3,3,4)
plt.plot(x,y1,y2,y3,y4,"g--")
plt.subplot(3,3,5)
plt.plot(x,y1,y2,y3,y4,"r--")
plt.subplot(3,3,6)
plt.plot(x,y1,y2,y3,y4,"b--")
plt.subplot(3,3,7)
plt.plot(x,y1,y2,y3,y4,"or--")
plt.subplot(3,3,8)
plt.plot(x,y1,y2,y3,y4,"r--")
plt.subplot(3,3,9)
plt.plot(x,y1,y2,y3,y4,"b--")
```
