Plot using sns.relplot, line plot, box plot on fmri data set

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fmri = sns.load_dataset("fmri")
```

```python
fmri.head(5)
```

```python
#relplot
sns.relplot(x = "timepoint", y = "signal", data = fmri, style = "event", hue = "region")
```

```python
#line plot
sns.lineplot(x = "subject", y = "event", data = fmri, hue = "region", style = "event")
```

```python
sns.catplot(x = "region", y = "signal", kind = "box", data = fmri, hue = "region", dodge = True)
```
